```python
# Use the function getTreeVars to a send a list of all the variables in the model
def getTreeVars( TREE, varNames ) :
    tree_ = TREE.tree_
    varName = [ varNames[i] if i != _tree.TREE_UNDEFINED else "undefined!" for i in tree_.feature ]

    nameSet = set()
    for i in tree_.feature :
        if i != _tree.TREE_UNDEFINED :
            nameSet.add( i )
    nameList = list( nameSet )
    parameter_list = list()
    for i in nameList :
        parameter_list.append( varNames[i] )
    return parameter_list
```

```python
# Create a text file for graphviz
feature_cols = list( X.columns.values ) 
vars_tree_flag = getTreeVars( dl_tree, feature_cols )  

# List the variables included in the decision tree that predict loan default
for i in vars_tree_flag :
    print(i)
```

```python
# Use the function getEnsembleTreeVars to only keep variables where their importance was above average
def getEnsembleTreeVars( ENSTREE, varNames ) :
    importance = ENSTREE.feature_importances_
    index = np.argsort(importance)
    theList = []
    for i in index :
        imp_val = importance[i]
        if imp_val > np.average( ENSTREE.feature_importances_ ) :
            v = int( imp_val / np.max( ENSTREE.feature_importances_ ) * 100 )
            theList.append( ( varNames[i], v ) )
    theList = sorted(theList,key=itemgetter(1),reverse=True)
    return theList
```

```python
feature_cols = list( X.columns.values )
vars_tree_flag = getEnsembleTreeVars( dl_tree, feature_cols )

# List the variables included in the Random Forest that predict loan default
for i in vars_tree_flag :
    print( i )
```
